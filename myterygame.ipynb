```python
import random

def intro():
    """Displays the game introduction."""
    print("Welcome to the Whispering Woods!")
    print("You find yourself at the edge of an ancient forest, shrouded in an eerie mist.")
    print("A weathered signpost points towards two paths:")
    print("1. The Sunlit Path")
    print("2. The Moonlit Trail")
    print("Which path will you choose?")

def choose_path():
    """Allows the player to choose a path."""
    while True:
        try:
            choice = int(input("Enter 1 for the Sunlit Path or 2 for the Moonlit Trail: "))
            if choice in (1, 2):
                return choice
            else:
                print("Invalid choice. Please enter 1 or 2.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def encounter(path):
    """Handles the player's encounter based on the chosen path."""
    if path == 1:
        print("\nYou follow the Sunlit Path, bathed in warm sunlight.")
        print("You encounter a mischievous sprite playing hide-and-seek amongst the trees.")
        sprite_mood = random.choice(["friendly", "grumpy"])
        if sprite_mood == "friendly":
            print("The sprite, delighted to see you, offers you a magical berry that grants temporary invisibility.")
            print("Do you accept the berry?")
            if input("Yes/No: ").lower() == "yes":
                print("You accept the berry and gain invisibility!")
            else:
                print("You decline the berry.")
                print("The sprite looks disappointed and disappears.")
            treasure_hunt()
        else:
            print("The grumpy sprite plays a mischievous trick on you, temporarily disorienting you.")
            print("You spend a few minutes regaining your bearings.")
            find_clearing()
    elif path == 2:
        print("\nYou venture down the Moonlit Trail, the air growing colder.")
        print("You encounter a wise old owl perched on a gnarled branch.")
        owl_advice = random.choice(["seek the hidden spring", "avoid the crossroads", "trust your instincts"])
        print("The owl hoots and gives you cryptic advice: ", owl_advice)
        print("Do you heed the owl's advice?")
        if input("Yes/No: ").lower() == "yes":
            print("You remember the owl's words.")
        else:
            print("You dismiss the owl's advice as foolishness.")
        magical_creature()

def treasure_hunt():
    """Handles the treasure hunt encounter."""
    print("\nAs you continue along the path, you find a treasure map lying on the ground.")
    print("Do you follow the map to seek the hidden treasure?")
    if input("Yes/No: ").lower() == "yes":
        print("You follow the map and find a hidden chest filled with gold coins!")
        dangerous_crossing()
    else:
        print("You decide to ignore the map and continue on your way.")
        peaceful_glade()

def find_clearing():
    """Handles the clearing encounter."""
    print("\nAfter wandering through the forest, you find a hidden clearing.")
    print("In the center of the clearing is a mysterious stone altar.")
    print("Do you investigate the altar?")
    if input("Yes/No: ").lower() == "yes":
        print("You discover ancient runes on the altar and gain magical knowledge.")
        ancient_artifact()
    else:
        print("You decide not to investigate and continue your journey.")
        shadow_figure()

def magical_creature():
    """Handles the magical creature encounter."""
    print("\nAs you walk further, you encounter a magical creature trapped in a snare.")
    print("The creature begs for your help. Do you free it?")
    if input("Yes/No: ").lower() == "yes":
        print("The creature thanks you and grants you a special ability to understand animal languages.")
        hidden_cave()
    else:
        print("You ignore the creature's plea and move on.")
        enchanted_garden()

def dangerous_crossing():
    """Handles the dangerous crossing encounter."""
    print("\nYou come across a rickety bridge over a deep ravine. Do you cross it?")
    if input("Yes/No: ").lower() == "yes":
        print("You carefully cross the bridge and make it to the other side safely.")
    else:
        print("You decide not to cross and find another way around, but it takes more time.")

def peaceful_glade():
    """Handles the peaceful glade encounter."""
    print("\nYou find a peaceful glade with a small pond. Do you rest here for a while?")
    if input("Yes/No: ").lower() == "yes":
        print("You rest by the pond and feel rejuvenated.")
    else:
        print("You continue on your journey without stopping.")

def ancient_artifact():
    """Handles the ancient artifact encounter."""
    print("\nYou discover an ancient artifact hidden in the altar. Do you take it with you?")
    if input("Yes/No: ").lower() == "yes":
        print("You take the artifact, feeling its power course through you.")
    else:
        print("You leave the artifact undisturbed.")

def shadow_figure():
    """Handles the shadow figure encounter."""
    print("\nYou notice a shadowy figure following you. Do you confront it?")
    if input("Yes/No: ").lower() == "yes":
        print("You confront the figure, and it turns out to be a lost traveler seeking help.")
    else:
        print("You ignore the figure and continue on your way.")

def hidden_cave():
    """Handles the hidden cave encounter."""
    print("\nYou find a hidden cave behind a waterfall. Do you explore it?")
    if input("Yes/No: ").lower() == "yes":
        print("You find hidden treasures inside the cave.")
    else:
        print("You decide not to explore the cave and move on.")

def enchanted_garden():
    """Handles the enchanted garden encounter."""
    print("\nYou stumble upon an enchanted garden filled with glowing flowers. Do you pick some flowers?")
    if input("Yes/No: ").lower() == "yes":
        print("You pick some flowers and feel their magic energize you.")
    else:
        print("You leave the flowers and continue on your journey.")

def ending():
    """Displays the game ending based on player choices."""
    print("\nYou reach the heart of the Whispering Woods.")
    print("You find a hidden clearing with a shimmering pool of water.")
    print("You have reached your destination.")

def play_game():
    """Plays the entire game."""
    intro()
    path = choose_path()
    encounter(path)
    ending()

if __name__ == "__main__":
    play_game()
```
